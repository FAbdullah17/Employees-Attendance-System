```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.preprocessing import LabelEncoder, Normalizer
from keras.models import load_model
import os
import cv2
from tqdm import tqdm
import pickle
from keras_facenet import FaceNet
```

```python
X_train = np.load("C:/Users/Fahad's WorkStation/Programs/PF/Employees-Attendance-System/data/processed/X_train.npy")
X_test = np.load("C:/Users/Fahad's WorkStation/Programs/PF/Employees-Attendance-System/data/processed/X_test.npy")
y_train = np.load("C:/Users/Fahad's WorkStation/Programs/PF/Employees-Attendance-System/data/processed/y_train.npy")
y_test = np.load("C:/Users/Fahad's WorkStation/Programs/PF/Employees-Attendance-System/data/processed/y_test.npy")
print(f"Train samples: {X_train.shape}, Test samples: {X_test.shape}")
```

```python
def get_embedding(model, face_pixels):
    face_pixels = face_pixels.astype('float32')
    mean, std = face_pixels.mean(), face_pixels.std()
    face_pixels = (face_pixels - mean) / std
    samples = np.expand_dims(face_pixels, axis=0)
    embedding = model.predict(samples)
    return embedding[0]

# Generate embeddings
X_train_embed = np.array([get_embedding(model, img) for img in tqdm(X_train)])
X_test_embed = np.array([get_embedding(model, img) for img in tqdm(X_test)])
```

```python
# Normalize embeddings
in_encoder = Normalizer(norm='l2')
X_train_norm = in_encoder.transform(X_train_embed)
X_test_norm = in_encoder.transform(X_test_embed)

# Encode labels
out_encoder = LabelEncoder()
out_encoder.fit(y_train)
y_train_enc = out_encoder.transform(y_train)
y_test_enc = out_encoder.transform(y_test)
```
